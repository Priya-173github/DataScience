```python
import pandas as pd
df = pd.read_excel('myfile.xlsx')
df.head()
```

String splitted upto first '|'

```python
df['Title'] = df['Title'].str.split('|').str[0]
df.head()
```

Removing views word from all the rows

```python
df['Views'] = df['Views'].str.replace('views', '')
df.head()
```

Convert Views into original numbers

```python
def convertNumbers(x):
    if 'K' in x:
        return float(x.replace('K',''))*1000
    elif 'M' in x:
        return float(x.replace('M', ''))*1000000
    else:
        return float(x)
```

```python
df['Views'] = df['Views'].apply(convertNumbers)
```

```python
df.head()
```

```python
df['Duration'] = df['Duration'].str.replace("\n", "");
df.head()
```

Converting time into minutes

```python
def ToMin(x):
    parts = list(map(int, x.split(':')))
    if len(parts) == 3:
        h,m,s = parts
    elif len(parts) == 2:
        h,m,s = 0, parts[0], parts[1]
    else:
        h,m,s = 0, 0, parts[0]
    minutes = h*60 + m + s/60
    return minutes
```

```python
df['Duration'] = df['Duration'].apply(ToMin)
df.head()
```

```python
df
```

```python
def classify(x):
    if x < 30:
        return 'Short'
    elif 30 <= x <= 60:
        return 'Medium'
    elif 60 < x <= 120:
        return 'Long'
    else:
        return 'Very Long'
```

```python
df['Duration Category'] = df['Duration'].apply(classify)
```
